# Molecule normalization using molvs

A quick tour on the default normalization / standardization methods of molvs.

```python
from rdkit import Chem
from molvs.standardize import Standardizer
from molvs.fragment import FragmentRemover

standardizer = Standardizer()
def_molecule = Chem.MolFromSmiles('[Na]OC(=O)c1ccc(C[S+2]([O-])([O-]))cc1')
print('[Na]OC(=O)c1ccc(C[S+2]([O-])([O-]))cc1   (not normalized)')
def_molecule
```

```python
standardized_molecule = standardizer.standardize(def_molecule)
print(f"{Chem.MolToSmiles(standardized_molecule)}   (normalized)")
standardized_molecule
```

```python
frag = FragmentRemover()
cleaned_molecule = frag.remove(standardizer.standardize(standardized_molecule))
print(f"{Chem.MolToSmiles(cleaned_molecule)}    (normalized+metals removed)")
cleaned_molecule
```
